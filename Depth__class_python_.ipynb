# Class Attributes 
  * class state
  * class var

```python
class TestClass:
  lang = "Arabic"
  number_speakers = 500000000
```

```python
#since class TestClass is an object we can use getattr
# getattr(object_name,attribute_name,optional_defualt) 
getattr(TestClass,'lang')
```

```python
# equivalent 
TestClass.lang
```

```python
# Also we have a setter the same as getattr
setattr(TestClass,"lang","English")
# we can add another attr 
setattr(TestClass,"number_learners",9e7)
# similar to 
TestClass.number_learners = 9e7
```

```python
getattr(TestClass,'lang')
```

```python
# Stroe in dict 
TestClass.__dict__
```

```python
# Remove attr 
# use delattr (objec_symbol,attr_name)
delattr(TestClass,'number_learners')
# or since we are working with dict we can use the del keyword  
del TestClass.number_speakers
```

```python
TestClass.__dict__
```

# Methods --> 
* An actual object type in python like function is calleble 
* the different method is bound to some object while function does not 
* this object passed to the method as its first parameter 

```python
test1.say_hello() # it is bound to the object test1 
```

```python
TestClass.say_hello(test1)
```

```python
class TestClass:
  lang = "Arabic"
  number_speakers = 500000000
  def say_hello(self):
    print("Hello this is my world")
```

```python
test1 = TestClass()
```

```python
test1.say_hello() # it is bound to the object test1 
```

```python
TestClass.say_hello(test1)
```

```python
# we said the method also an object has attributes 
# __self__ --> the instance method is bound to  
# __func__ --> the original functinon --> defined in the class
```

```python
test1.say_hello.__func__
```

```python
test1.say_hello.__self__
```
